```python
import random
print("Run 1: ", [random.random() for _ in range(3)])
print("Run 2: ", [random.random() for _ in range(3)])
```

```python
import random

random.seed(42)
print("Run 1: ", [random.random() for _ in range(3)])

random.seed(42)
print("Run 2: ", [random.random() for _ in range(3)])
```

```python
import numpy as np

print("Run 1: ", np.random.rand(3))
print("Run 2: ", np.random.rand(3))
```

```python
import numpy as np

np.random.seed(42)
print("Run 1: ", np.random.rand(3))

np.random.seed(42)
print("Run 2: ", np.random.rand(3))
```

```python
import torch

print("Run 1: ", torch.rand(3))
print("Run 2: ", torch.rand(3))
```

```python
import torch

torch.manual_seed(42)
print("Run 1: ", torch.rand(3))

torch.manual_seed(42)
print("Run 2: ", torch.rand(3))
```
